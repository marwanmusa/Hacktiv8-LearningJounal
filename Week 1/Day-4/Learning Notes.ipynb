```python
#drop column
nba.drop(labels=['elo_i', 'elo_n', 'opp_elo_i', 'opp_elo_n'], axis=1, inplace=True)
```

```python
nba.head()
```

```python
# siapkan list kosong
# definisikan args dari for loop
# simpan hasil ke salam list kosong
# jika sudah selesai convert list menjadi kolom
# kolom pts -> nba.iloc[:, 10]
# kolom opp_pts -> nba.iloc[:, 14]

hasil =  []
for idx in range(0,len(nba)):
    selisih = nba.iloc[idx, 10] - nba.iloc[idx, 14]
    hasil.append(selisih)

hasil
    
```

```python
nba['pts_substract'] = hasil
```
