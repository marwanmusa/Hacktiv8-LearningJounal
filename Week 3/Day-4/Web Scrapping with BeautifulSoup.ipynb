## *Web Scrapping with BeautifulSoup*

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
response = requests.get('https://news.ycombinator.com/').text
print(response)
```

```python
soup = BeautifulSoup(response, 'lxml')

title = soup.find('a', class_= 'titlelink').text
print(title)
score = soup.find('span', class_ = 'score').text
print(score)
```

```python
title2 = soup.find('a', class_='titlelink')
info = title2['href']
info
```

find all

```python
df1 = pd.DataFrame()
df2 = pd.DataFrame()
df3 = pd.DataFrame()

df1['title'] = [title.text for title in soup.find_all('a', class_ = 'titlelink')]
df2['score'] = [int(score.text.split()[0]) for score in soup.find_all('span', class_ = 'score')]
df3['info'] = [info['href'] for info in soup.find_all('a', class_ = 'titlelink')]

df_final = pd.concat([df1, df2, df3], axis=1)
df_final
```
